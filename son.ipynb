```python
import time
import datetime
import pandas as pd
import random
print("Restorana hosgeldiniz")
menu = pd.DataFrame({"Yemekler": ["Pizza","Makarna","Çorba","Pilav"],
                     "Çeşit : 1" : ["Margarita Pizza: 50 TL","Spagetti Makarna: 17 TL","Mercimek: 10 TL","Tavuklu Pilav: 21 TL"],
                     "Çeşit : 2" : ["Etli Pizza: 40 TL", "Kelebek Makarna: 15 TL","Yayla Çorbası: 10 TL","Bulgur Pilavı: 21 TL"],
                     "Çeşit : 3" : ["Veggie Pizza: 40 TL","Boncuk Makarna: 14 TL","Tarhana Çorbası: 14 TL","Nohutlu Pilav: 17 TL"]})
print(menu)
print("*"*20)
menu_icecek = pd.DataFrame({"İçecekler" : ["Coco Cola","Fanta","IceTea","Meyve Suyu","Ayran"],
                                           "Fiyat" : [20,18,17,35,15]})
print(menu_icecek)
toplam_fiyat = 0

class Menu:
  def __init__(self,isim,fiyat ):
    self.isim = isim
    self.fiyat = fiyat

class Pizza(Menu):
    def __init__(self, isim, fiyat, icindekiler,boy,kenar,sos):
      super().__init__(isim, fiyat)
      self.icindekiler = icindekiler
      self.boy = boy
      self.kenar = kenar
      self.sos = sos
    def SiparisGoster(self):
        print(f" Sipariş: {self.isim}, Fiyat:{self.fiyat}, İçindekiler: {self.icindekiler}, Sosu: {self.sos}")


    def siparis(self):
        s = input("siparişinizde değişiklik yapmak istiyor musunuz: ").lower()
        if s == "evet":
            e = input(("Eklemek ıstedıgınız urun: ")).lower()
            c = input(("Cıkartmak ıstedıgınız urun: ")).lower()
            if e == "yok":
                pass
            else:       
                self.icindekiler.append(e)
                print("ozel isteginiz kaydedildi")
            if c == "yok":
                pass
            else: 
                self.icindekiler.remove(c)
                print("ozel isteginiz kaydedildi")
        
  
  
class Çorba(Menu):
    def __init__(self, isim, fiyat):
      super().__init__(isim, fiyat)
    def SiparisGoster(self):
      print(f"Sipariş: {self.isim}, Fiyat:{self.fiyat}, Baharat : {baharat}")

class Makarna(Menu):
    def __init__(self, isim, fiyat):
      super().__init__(isim, fiyat)

    def SiparisGoster(self):
        print(f"Sipariş: {self.isim}, Sos: {sos} Fiyat:{self.fiyat}")           

class Pilav(Menu):
    def __init__(self, isim, fiyat,cesit):
      super().__init__(isim, fiyat )
      self.cesit = cesit
    def SiparisGoster(self):
        print(f"Sipariş: {self.isim}, Cesit: {self.cesit} Fiyat:{self.fiyat}")

class Icecek(Menu):
    def _init_(self, isim, fiyat):
      super()._init_(isim, fiyat )
    def SiparisGoster(self):
      print(f"Sipariş: {self.isim}, Boyut: {b} Fiyat:{self.fiyat}")

margarita= 0
veggi = 0
etli=0
mercimek= 0
yayla= 0
tarhana= 0
spagetti =0
kelebek= 0
boncuk=0
nohutlu=0
tavuklu=0
bulgur=0
cola=0
ayran=0
icetea=0
fanta=0
meyvesuyu=0



l1 = ["mozarella", "domates"]
l2 = ["mozarella", "domates", "biber", "patlican"]
l3 = ["mozarella", "domates", "sucuk"]
menusozluk = {'margarita': [50, l1], "etli" : [40,l2], "veggie" : [40, l3], #hepsini duzelt
'spagetti' : 17, "kelebek" : 15,  "boncuk" : 14,
'sercimek': 10, "yayla" : 10, "tarhana" : 14,
'tavuklu': 21, "bulgur" : 21,  "nohutlu" : 17}
imenusozluk = {'cocacola': 20, "ayran" : 10, "icetea" : 40, "fanta" : 15, "meyvesuyu": 10}
siparisler = []
while True:
    
        urun = input("Sipariş etmek istediğiniz urunu girin: (Pizza,Makarna,Çorba,Pilav,İçecek) " ).lower()
        if urun == "pizza":
           
            sayi = int(input("Lütfen kac adet pizza siparis edeceginizi giriniz: "))
            
            for i in range(sayi):
              siparis = input("Lutfen pizzanizi seciniz (margarita, etlipizza, veggiepizza), bosluk birakmayiniz" .format(i+1) )
              boy =input("Lütfen boyutunu giriniz:")
              kenar = input("Lütfen kenarı giriniz:")
              sos = input("Lutfen sos giriniz: ")
              if siparis == 'margarita':
                margarita +=1
                pizza = Pizza(isim= siparis, fiyat = menusozluk[siparis][0] , icindekiler=menusozluk[siparis][1] , boy = boy, kenar = kenar, sos = sos)
                print("Icındekıler: ", l1)
                pizza.siparis()
                pizza.SiparisGoster()
                siparisler.append(siparis)
                siparisler.append(menusozluk[siparis][0])
                toplam_fiyat += menusozluk[siparis][0]
                print("-"*40)

              elif siparis == 'etli':
                etli +=1
                pizza = Pizza(isim= siparis, fiyat = menusozluk[siparis][0] , icindekiler=menusozluk[siparis][1] , boy = boy, kenar = kenar, sos = sos)
                print("Icındekıler: ", l2)
                pizza.siparis()
                pizza.SiparisGoster()
                siparisler.append(siparis)
                siparisler.append(menusozluk[siparis][0])
                toplam_fiyat += menusozluk[siparis][0]
                print("-"*40)
            
              elif siparis == 'veggie':
                veggi +=1
                pizza = Pizza(isim= siparis, fiyat = menusozluk[siparis][0] , icindekiler=menusozluk[siparis][1] , boy = boy, kenar = kenar, sos = sos)
                print("Icındekıler: ", l3)
                pizza.siparis()  
                pizza.SiparisGoster()
                siparisler.append(siparis)
                siparisler.append(menusozluk[siparis][0])
                toplam_fiyat += menusozluk[siparis][0]
                print("-"*40)


        elif urun == "corba":
          print("06.00-12.00 arasında çorba servisimiz vardır, maalesef bu saatler dışında servis yoktur")
          try:
            now = datetime.datetime.now() 
             
            if 6 <= now.hour < 12:
               print(f"Saat : {now.hour}.00 , {urun} serivisi vardır.")
               x =  int(input("Lütfen kac adet çorba siparis edeceginizi giriniz: "))
               
               for c in range(x):
                siparis = input(f"Lutfen corba seciminizi gırınız (mercimek,yayla,tarhana)" )
                by = int(input("Baharat istiyorsanız lütfen 1'e , baharat istemiyorsanız lütfen 0'a basın."))
               if by == 1:
                  baharat = input("Lütfen baharatı giriniz:")
                  print(f"{baharat} ekleniyor")
               elif by == 0:
                  pass
               corba =Çorba(isim=siparis, fiyat= menusozluk[siparis])
               corba.SiparisGoster()
               siparisler.append(siparis)
               siparisler.append(menusozluk[siparis])
               toplam_fiyat += menusozluk[siparis]
               print("-"*40)
            else:
               print(f"Saat : {now.hour}.{now.minute}, {urun} servisi yoktur.")
               pass
          except Exception as e:
            print(f"{urun} servisinde bir hata oluştu ve sebebi:{e}")
          if siparis == "mercimek":
            mercimek +=1
          elif siparis == "yayla":
            yayla +=1
          elif siparis == "tarhana":
             tarhana += 1      


        elif urun == "makarna":
          
          adet =  int(input("Lütfen kac adet makarna siparis edeceginizi giriniz: "))
          
          for b in range(adet):
            siparis = input("Lutfen makarna sıparısinizi gırınız (spagetti, kelebek, boncuk)" .format(b+1) )
            sos = input("Lütfen sos giriniz:").lower()
            if sos == "yok":
               continue
            else:
              print(f"{sos} ekleniyor")
            makarna =Makarna(isim=siparis, fiyat= menusozluk[siparis])
            makarna.SiparisGoster()
            siparisler.append(siparis)
            siparisler.append(menusozluk[siparis])
            toplam_fiyat += menusozluk[siparis]
            print("-"*40)

            if siparis == "spagetti":
              spagetti += 1
            elif siparis == "kelebek":
              kelebek += 1
            elif siparis == "boncuk":
              boncuk += 1 

        elif urun == "pilav":
            s =  int(input("Lütfen kac adet pilav siparis edeceginizi giriniz: "))
           
            for a in range(s):
              siparis = input("Lutfen pilav sıparısinizi gırınız (nohutlu, tavuklu, bulgur)" .format(a+1) )
              pilav =Pilav(isim=siparis, fiyat= menusozluk[siparis], cesit =siparis )
              pilav.SiparisGoster()
              siparisler.append(siparis)
              siparisler.append(menusozluk[siparis])
              toplam_fiyat += menusozluk[siparis]
              print("-"*40)

              if siparis == "nohutlu":
                nohutlu +=1
              elif siparis == "tavuklu":
                tavuklu +=1
              elif siparis == "bulgur":
                bulgur += 1

        elif urun == "icecek":
         ic =  int(input("Lütfen kac adet içecek siparis edeceginizi giriniz: "))
         for w in range(ic):
            siparis = input("Lutfen icecek sıparısinizi gırınız (cocacola, ayran, icetea, fanta, meyvesuyu)" .format(w+1) )
            b = input("Lütfen içeceğinizin boyutunu giriniz: ")
            
            icecek = Icecek(isim=siparis, fiyat= imenusozluk[siparis])
            icecek.SiparisGoster()
            siparisler.append(siparis)
            siparisler.append(imenusozluk[siparis])
            toplam_fiyat += imenusozluk[siparis]
            print("-"*40)
            
            if siparis == "cocacola":
              cola +=1
            elif siparis == "ayran":
              ayran +=1
            elif siparis == "icetea":
              icetea +=1
            elif siparis == "fanta":
              fanta +=1
            elif siparis == "meyvesuyu":
              meyvesuyu +=1
                
        else:
            print("girdiğiniz ürün yoktur, lütfen tekrar seçiniz")
            continue
        
        
        urun_ekleme = input("başka bir urun eklemek istiyor musunuz: (Evet-Hayır)").lower()
        if urun_ekleme == "evet":
            continue
        else:
            pass
        kontrol = input("Yaptığınız işlemler doğru mu? (Evet-Hayır)").lower()
        if kontrol == "evet":
            break

isim = input("Isminizi ve soyisminizi giriniz: ")
print("Toplam fiyat: ", toplam_fiyat)

tatlilar = ["Baklava","İrmik helvası","Şekerpare","Künefe","Profiterol","Kazandibi","Lokum","Kadayıf","Trileçe","Çikolatalı lava kek"]
sansizlik = [0,0,0]
tatli_ikram_sansi = tatlilar + sansizlik
sonuc = random.choice(tatli_ikram_sansi)
print("Sizi bir sürprizimiz ile tanıştırıyoruz, eğer şanslıysanız size tatlı ikram edeceğiz")
time.sleep(5)

if sonuc == 0:
   print("Bugün şansınızı başka bir yerde kullandınız herhalde, maalesef tatlı alma şansını kaybettiniz.")
else:
   print(f"Tebrikler tatlı almaya hak kazandınız ve tatlınız : {sonuc}'dır")

for i in range(5,35,5):
  if datetime.datetime.now().day == i :
    print("Sizi bir sürprizimiz ile daha tanıştırıyoruz, Bizi her ayın 5.,10.,15.,20.,25. ve 30. günlerinde tercih ederseniz ki siz ettiniz %20 tüm ürünlerde indirim yaparak uğurluyoruz")
    print(f"Normalde toplam fiyat: {toplam_fiyat}")
    toplam_fiyat = toplam_fiyat*80/100
    print(f"Bugüne özel Toplam Fiyat : {toplam_fiyat}")
  else:
    pass


tur = input("Lütfen ödeme türünüzü seçiniz. Kart mı? Nakit mi?").lower()
if tur == "nakit":
    print("Ödemeniz nakit alınacaktır.")
    verilen_ucret = int(input("Hesabı ödemek için ne kadar para verdiğiniz giriniz."))
    para_ustu = verilen_ucret - toplam_fiyat
    print(f"Para Üstü:  {para_ustu}")
elif tur == "kart":
    print("Ödemeniz karttan alınacaktır.")
    print("dıt")
    time.sleep(1)
    print("Karttan ödeme işleminiz onaylandı")
    print(f"Karttan Çekilen Ücret : {toplam_fiyat}")

s_siparisler = ['{}\t'.format(i) for i in siparisler]
with open("fis.txt", "w+", encoding = "utf-8") as f: #deneme.txt dosyasında pizza ve diğer ürünlerin maliyetini varsayım şeklinde girebiliriz. Çünkü her ürünün farklı işletmelere maliyeti farklı olur.
    f.write(str(time.ctime(time.time())))
    lines = [f"\n RESTORANT FİŞİ\n İsim/Soyisim: {isim}\n ToplamFiyat: {toplam_fiyat}\n ÖdemeTürü: {tur}\n "] 
    f.writelines(lines)
    f.writelines(s_siparisler)

soru = input("Geri bildirim anketimize katılmak ister misiniz? ")

if soru == "hayir":
  print("İyi günler dileriz")
  

elif soru == "evet":
  yemekler = int(input("Lütfen yemeklerimizi beş üzerinden değerlendiriniz: "))
  servis = int(input("Lütfen servisimizi beş üzerinden değerlendiriniz: "))
  restorant= int(input("Lütfen restorantımızı beş üzerinden değerlendiriniz: "))
  print("Geri dönüş sağladığınız için teşekkür eder, iyi günler dileriz.")
data= {"Yemekler": [{yemekler}],
        "Servis" : [{servis}],
        "restorant": [{restorant}]}

df = pd.DataFrame(data)
df.to_excel("/Users/ADA/Desktop/samurai.xlsx")
```
